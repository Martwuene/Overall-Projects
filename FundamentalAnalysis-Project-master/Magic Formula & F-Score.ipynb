### 1. Import libraries

```python
# import library
from information import magic_fscore
from information import fscore_magic
from information import financial_indicators
import information
import data
import pandas as pd
import numpy as np
```

### 2. Data Loading

```python
# extracting data for the analysis
stock_data_info = pd.read_excel('JSE Total Companies List_rev1.xlsx', usecols=['Code', 'Company', 'Sector'],header=1)
```

```python
# data extraction reporting
balance_sheet, income_statement, cashflow_statement, tickers = data.report(stock_data_info.Code)
```

### 3.  Ranking using Greenblatt's Magic Formula & F-Score




<ins>__Johannesburg Stock Exchange Market Capitalization:__</ins> 

- __Small Cap:__ Market cap is below R 1 Billion. 
- __Medium Cap:__ Market cap of between R1 Billion and R10 Billion. 
- __Large Cap:__ Market cap above R10 Billion.

```python
# displaying required financial indicators
financial_data = financial_indicators(balance_sheet, income_statement, cashflow_statement, ticker=tickers)

# filling NaN values with zeros
financial_data = financial_data.fillna(0)

# displaying required financial metric
financial_metrics = information.financial_metrics(df=financial_data, ticker=tickers)
```

```python
# displaying stocks that are listed by magic formula to f-score [2022 - 2020]
magic_fscore(financial_metrics, financial_data, stock_data_info, 2022)
```

```python
# displaying stocks that are listed by f-score to magic formula [2022 - 2020]
fscore_magic(financial_metrics, financial_data, stock_data_info, 2022)
```
